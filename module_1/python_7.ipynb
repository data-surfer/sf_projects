# 7.11 Объединяем выгрузки

```python
# Объединить набор файлов в один датафрейм, нам поможет библиотека os

import os
```

```python
# Для получения списка файлов воспользуемся методом listdir. 
# В качестве аргумента указываем папку 'data'

files = os.listdir('data')
files
```

```python
# Создайте на основе списка files новый список data, в который поместите 
# только файлы, содержащие в названии "txt"


data = [word for word in files if '.txt' in word]
```

```python
data
```

# 7.13 Склеивание датафреймов

```python
# содержимое этих файлов, собрать в одном датафрейме. 
# Это можно сделать используя метод concatenate

# создадим пустой датафрейм data с нужными названиями столбцов
data = pd.DataFrame(columns = ['userId', 'movieId', 'rating', 'timestamp'])
```

```python
# Для каждого имени файла filename будем записывать его содержимое в датафрейм temp

# Напишите цикл, который собирает содержимое файлов папки data в единый датафрейм data.

for filename in files:
    temp = pd.read_csv(os.path.join('data', filename), 
                       names = ['userId', 'movieId', 'rating', 'timestamp'] )
    data = pd.concat([data, temp])
```
